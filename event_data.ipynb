```python
import pandas as pd
import requests
import time
from decimal import Decimal, ROUND_HALF_UP

# --- CONFIGURATION ---
HEADERS = {
    'User-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36'
}
INPUT_FILE = "full_histories_from_28_markets.csv"
OUTPUT_FILE = "trader_summary_from_resolved_markets.csv"

def get_market_resolution(condition_id):
    """
    Fetches market details to determine if it is resolved and what the outcome was.
    Returns 'YES', 'NO', 'INVALID', or 'ACTIVE'.
    """
    url = f"https://gamma-api.polymarket.com/events/{condition_id}"
    try:
        response = requests.get(url, headers=HEADERS, timeout=30)
        response.raise_for_status()
        market_data = response.json()
        
        if market_data.get('closed') and market_data.get('resolved'):
            for outcome in market_data.get('outcomes', []):
                if outcome.get('won'):
                    if outcome.get('name', '').lower() in ['yes', 'long']:
                        return 'YES'
                    elif outcome.get('name', '').lower() in ['no', 'short']:
                        return 'NO'
            return 'INVALID'
        else:
            return 'ACTIVE'
    except requests.exceptions.RequestException:
        return None

def main():
    print(f"--- Step 1: Reading and processing trades from {INPUT_FILE} ---")
    try:
        trades_df = pd.read_csv(INPUT_FILE)
        trades_df['Trade_Size'] = pd.to_numeric(trades_df['Trade_Size'], errors='coerce')
        trades_df['Trade_Price'] = pd.to_numeric(trades_df['Trade_Price'], errors='coerce')
        trades_df.dropna(subset=['Trade_Size', 'Trade_Price', 'Trader_Wallet', 'Market_Condition_ID'], inplace=True)
        print(f"‚úÖ Found {len(trades_df)} valid trades to analyze.")
    except FileNotFoundError:
        print(f"‚ùå ERROR: The file '{INPUT_FILE}' was not found.")
        return

    unique_markets = trades_df['Market_Condition_ID'].unique()
    print(f"\n--- Step 2: Found {len(unique_markets)} unique markets. Fetching their resolution status... ---")

    market_resolutions = {}
    for i, market_id in enumerate(unique_markets):
        print(f"  ({i+1}/{len(unique_markets)}) Fetching status for {market_id[:10]}...")
        status = get_market_resolution(market_id)
        if status:
            market_resolutions[market_id] = status
        time.sleep(1)

    resolved_markets_count = sum(1 for status in market_resolutions.values() if status in ['YES', 'NO', 'INVALID'])
    print(f"‚úÖ Found {resolved_markets_count} markets that are resolved.")

    print("\n--- Step 3: Calculating Profit/Loss for each trader... ---")
    pnl_records = []
    grouped_by_trader_and_market = trades_df.groupby(['Trader_Wallet', 'Market_Condition_ID'])

    for (trader, market_id), group in grouped_by_trader_and_market:
        resolution = market_resolutions.get(market_id)
        if not resolution or resolution not in ['YES', 'NO']:
            continue

        net_cash_flow = Decimal('0')
        net_yes_shares = Decimal('0')
        net_no_shares = Decimal('0')

        for _, row in group.iterrows():
            size = Decimal(str(row['Trade_Size']))
            price = Decimal(str(row['Trade_Price']))
            trade_value = size * price
            
            if row['Trade_Side'] == 'BUY':
                net_cash_flow -= trade_value
            elif row['Trade_Side'] == 'SELL':
                net_cash_flow += trade_value
            
            if row['Trade_Outcome'] == 'Yes':
                if row['Trade_Side'] == 'BUY': net_yes_shares += size
                else: net_yes_shares -= size
            elif row['Trade_Outcome'] == 'No':
                if row['Trade_Side'] == 'BUY': net_no_shares += size
                else: net_no_shares -= size

        payout_value = Decimal('0')
        if resolution == 'YES':
            payout_value = net_yes_shares * Decimal('1')
        elif resolution == 'NO':
            payout_value = net_no_shares * Decimal('1')
            
        market_pnl = payout_value + net_cash_flow
        pnl_records.append({'Trader_Wallet': trader, 'Market_Condition_ID': market_id, 'Pnl': market_pnl})

    pnl_df = pd.DataFrame(pnl_records)

    print("\n--- Step 4: Aggregating results for the final summary... ---")
    predictions = trades_df.groupby('Trader_Wallet').size().reset_index(name='Predictions')

    # --- FIX STARTS HERE ---
    if not pnl_df.empty:
        trader_summary = pnl_df.groupby('Trader_Wallet').agg(
            Profit_Loss=('Pnl', 'sum'),
            Biggest_Win=('Pnl', lambda x: x[x > 0].max())
        ).reset_index()
        trader_summary['Biggest_Win'].fillna(0, inplace=True)
        final_summary_df = pd.merge(predictions, trader_summary, on='Trader_Wallet', how='left')
    else:
        # If no P/L data exists, create a summary with 0s
        print("‚ö†Ô∏è No P/L data was calculated as no resolved markets were found. P/L will be 0.")
        final_summary_df = predictions
        final_summary_df['Profit_Loss'] = 0.0
        final_summary_df['Biggest_Win'] = 0.0
    # --- FIX ENDS HERE ---
    
    final_summary_df.fillna(0, inplace=True)

    final_summary_df['Profit_Loss'] = final_summary_df['Profit_Loss'].apply(lambda x: float(Decimal(x).quantize(Decimal('.01'), rounding=ROUND_HALF_UP)))
    final_summary_df['Biggest_Win'] = final_summary_df['Biggest_Win'].apply(lambda x: float(Decimal(x).quantize(Decimal('.01'), rounding=ROUND_HALF_UP)))
    final_summary_df = final_summary_df.sort_values(by='Profit_Loss', ascending=False)

    print("\n--- Final Trader Summary ---")
    print(final_summary_df.to_string())
    
    final_summary_df.to_csv(OUTPUT_FILE, index=False)
    print(f"\n‚úÖ Final summary saved to {OUTPUT_FILE}")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import requests
import time
import json
import os
from decimal import Decimal, ROUND_HALF_UP

# --- CONFIGURATION ---
HEADERS = {
    'User-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36'
}
TRADES_FILE = "full_histories_from_28_markets.csv"
# --- THIS IS THE CORRECT FILE WITH FULL TOKEN IDS ---
MARKETS_FILE = "event_condition_token_ids.csv"
OUTPUT_FILE = "trader_summary_final.csv"
CACHE_FILE = "market_data_cache.json"

def get_market_data_by_token(yes_token_id):
    """Fetches market data using the YES token ID from the CLOB API."""
    # The token ID from the correct file should not need formatting
    url = f"https://clob.polymarket.com/markets/{yes_token_id}"
    try:
        response = requests.get(url, headers=HEADERS, timeout=30)
        response.raise_for_status()
        market_data = response.json()
        
        if isinstance(market_data, dict) and 'error' in market_data:
             return None
        
        if isinstance(market_data, list):
            market_data = market_data[0] if market_data else {}
            
        if market_data.get('closed', False):
            outcome_prices = market_data.get('outcome_prices', [])
            if len(outcome_prices) >= 2:
                try:
                    if float(outcome_prices[0]) == 1.0 and float(outcome_prices[1]) == 0.0:
                        return {'status': 'YES'}
                    if float(outcome_prices[0]) == 0.0 and float(outcome_prices[1]) == 1.0:
                        return {'status': 'NO'}
                except (ValueError, IndexError):
                    pass
        
        price_sources = [
            market_data.get('outcome_prices', [None])[0],
            market_data.get('last_trade_price'),
        ]
        for price in price_sources:
            if price is not None:
                try:
                    price_dec = Decimal(str(price))
                    if 0 < price_dec < 1:
                        return {'status': 'ACTIVE', 'price': price_dec}
                except Exception:
                    continue
        
        return {'status': 'ACTIVE', 'price': Decimal('0.5')}
        
    except (requests.exceptions.RequestException, json.JSONDecodeError):
        return None

def main():
    print(f"--- Step 1: Reading input files ---")
    try:
        markets_df = pd.read_csv(MARKETS_FILE, dtype=str) # Read all as strings to preserve precision
        trades_df = pd.read_csv(TRADES_FILE)
    except FileNotFoundError as e:
        print(f"‚ùå ERROR: File not found: {e.filename}")
        return

    trades_df['Trade_Size'] = pd.to_numeric(trades_df['Trade_Size'], errors='coerce')
    trades_df['Trade_Price'] = pd.to_numeric(trades_df['Trade_Price'], errors='coerce')
    trades_df.dropna(subset=['Trade_Size', 'Trade_Price', 'Trader_Wallet', 'Market_Condition_ID'], inplace=True)
    
    market_tokens = {row['condition_id']: row['yes_token_id'] for _, row in markets_df.iterrows()}

    print(f"\n--- Step 2: Fetching data for unique markets... ---")
    unique_markets = trades_df['Market_Condition_ID'].unique()
    market_data_map = {}
    if os.path.exists(CACHE_FILE):
        with open(CACHE_FILE, 'r') as f:
            market_data_map = json.load(f)
        print(f"  Loaded {len(market_data_map)} markets from cache.")

    for i, market_id in enumerate(unique_markets):
        if market_id in market_data_map:
            continue
        
        yes_token = market_tokens.get(market_id)
        if not yes_token:
            print(f"  ({i+1}/{len(unique_markets)}) Market {market_id[:10]}... Status: No Token ID")
            continue
            
        data = get_market_data_by_token(yes_token)
        status_str = "Error"
        if data:
            market_data_map[market_id] = data
            status_str = data.get('status', 'Unknown')
        
        print(f"  ({i+1}/{len(unique_markets)}) Market {market_id[:10]}... Status: {status_str}")
        time.sleep(1)

    with open(CACHE_FILE, 'w') as f:
        # Convert Decimal to string for JSON serialization
        json.dump({k: ({'status': v['status'], 'price': str(v['price'])} if 'price' in v else v) for k, v in market_data_map.items()}, f)
    print(f"\n‚úÖ Market data fetching complete. Cache saved.")

    print("\n--- Step 3: Calculating P/L and Positions Value... ---")
    results = []
    grouped = trades_df.groupby(['Trader_Wallet', 'Market_Condition_ID'])

    for (trader, market_id), group in grouped:
        market_info_str = market_data_map.get(market_id)
        if not market_info_str:
            continue
        # Convert price back to Decimal if it exists
        market_info = dict(market_info_str)
        if 'price' in market_info:
            market_info['price'] = Decimal(market_info['price'])

        net_cash, net_yes, net_no = Decimal('0'), Decimal('0'), Decimal('0')
        for _, row in group.iterrows():
            size, price = Decimal(str(row['Trade_Size'])), Decimal(str(row['Trade_Price']))
            value = size * price
            
            if row['Trade_Side'] == 'BUY': net_cash -= value
            else: net_cash += value
            
            if row['Trade_Outcome'] == 'Yes':
                if row['Trade_Side'] == 'BUY': net_yes += size
                else: net_yes -= size
            else:
                if row['Trade_Side'] == 'BUY': net_no += size
                else: net_no -= size
        
        status = market_info.get('status')
        if status in ['YES', 'NO']:
            payout = (net_yes * Decimal('1')) if status == 'YES' else (net_no * Decimal('1'))
            pnl = payout + net_cash
            results.append({'Trader_Wallet': trader, 'Pnl_Realized': pnl, 'Pnl_Unrealized': Decimal('0')})
        elif status == 'ACTIVE':
            price = market_info.get('price', Decimal('0.5'))
            market_value = (net_yes * price) + (net_no * (Decimal('1') - price))
            pnl = market_value + net_cash
            results.append({'Trader_Wallet': trader, 'Pnl_Realized': Decimal('0'), 'Pnl_Unrealized': pnl})

    results_df = pd.DataFrame(results)
    if results_df.empty:
        print("No results to process.")
        return

    print("\n--- Step 4: Aggregating final summary... ---")
    predictions = trades_df.groupby('Trader_Wallet').size().reset_index(name='Predictions')
    
    realized_summary = results_df.groupby('Trader_Wallet').agg(Pnl_Realized_Sum=('Pnl_Realized', 'sum')).reset_index()
    biggest_win_summary = results_df[results_df['Pnl_Realized'] > 0].groupby('Trader_Wallet').agg(Biggest_Win=('Pnl_Realized', 'max')).reset_index()
    unrealized_summary = results_df.groupby('Trader_Wallet').agg(Pnl_Unrealized_Sum=('Pnl_Unrealized', 'sum')).reset_index()

    final_summary = pd.merge(predictions, realized_summary, on='Trader_Wallet', how='left')
    final_summary = pd.merge(final_summary, biggest_win_summary, on='Trader_Wallet', how='left')
    final_summary = pd.merge(final_summary, unrealized_summary, on='Trader_Wallet', how='left')
    final_summary.fillna(0, inplace=True)
    
    final_summary.rename(columns={
        'Pnl_Realized_Sum': 'Profit_Loss_Realized',
        'Pnl_Unrealized_Sum': 'Positions_Value_Unrealized'
    }, inplace=True)
    
    for col in ['Profit_Loss_Realized', 'Biggest_Win', 'Positions_Value_Unrealized']:
        final_summary[col] = final_summary[col].apply(lambda x: float(Decimal(x).quantize(Decimal('.01'), rounding=ROUND_HALF_UP)))
    
    final_summary = final_summary.sort_values(by=['Profit_Loss_Realized', 'Positions_Value_Unrealized'], ascending=False)
    
    print("\n--- Final Trader Summary ---")
    print(final_summary.to_string())
    final_summary.to_csv(OUTPUT_FILE, index=False)
    print(f"\n‚úÖ Final summary saved to {OUTPUT_FILE}")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd

def analyze_trader_performance(filepath):
    """
    Analyzes trader performance from a CSV file of trade histories.

    Args:
        filepath (str): The path to the CSV file.

    Returns:
        pandas.DataFrame: A DataFrame with each trader's total P&L and biggest single-market win.
    """
    try:
        # Load the dataset
        df = pd.read_csv(filepath)
        print("Successfully loaded the CSV file.")
    except FileNotFoundError:
        print(f"Error: The file '{filepath}' was not found.")
        print("Please make sure the CSV file is in the same directory as this script.")
        return None

    # --- Data Cleaning and Preparation ---
    
    # Select and rename columns for clarity
    df = df[['Trader_Name', 'Market_Title', 'Trade_Side', 'Trade_Size', 'Trade_Price']].copy()
    
    # Convert 'Trade_Size' and 'Trade_Price' to numeric types, coercing errors
    df['Trade_Size'] = pd.to_numeric(df['Trade_Size'], errors='coerce')
    df['Trade_Price'] = pd.to_numeric(df['Trade_Price'], errors='coerce')

    # Drop rows where conversion might have failed
    df.dropna(subset=['Trade_Size', 'Trade_Price'], inplace=True)
    
    # --- Calculation ---

    # Calculate the value of each trade (cost or revenue)
    # BUY is a cost (negative), SELL is a revenue (positive)
    df['Signed_Value'] = df.apply(
        lambda row: -1 * row['Trade_Size'] * row['Trade_Price'] if row['Trade_Side'] == 'BUY'
        else row['Trade_Size'] * row['Trade_Price'],
        axis=1
    )
    
    # 1. Calculate Profit & Loss for each trader in each market
    market_pnl = df.groupby(['Trader_Name', 'Market_Title'])['Signed_Value'].sum().reset_index()
    market_pnl.rename(columns={'Signed_Value': 'Market_PnL'}, inplace=True)

    # 2. Find the biggest single win for each trader
    biggest_wins = market_pnl.loc[market_pnl.groupby('Trader_Name')['Market_PnL'].idxmax()]
    biggest_wins = biggest_wins[['Trader_Name', 'Market_Title', 'Market_PnL']]
    biggest_wins.rename(columns={'Market_Title': 'Biggest_Win_Market', 'Market_PnL': 'Biggest_Win_Amount'}, inplace=True)
    
    # 3. Calculate the total Profit & Loss for each trader across all markets
    total_pnl = df.groupby('Trader_Name')['Signed_Value'].sum().reset_index()
    total_pnl.rename(columns={'Signed_Value': 'Total_PnL'}, inplace=True)

    # --- Combine results into a summary table ---
    
    # Merge total P&L with the biggest wins
    summary_df = pd.merge(total_pnl, biggest_wins, on='Trader_Name', how='left')
    
    # Sort by total P&L for a ranked view
    summary_df = summary_df.sort_values(by='Total_PnL', ascending=False).reset_index(drop=True)

    return summary_df

if __name__ == "__main__":
    # Define the name of your input CSV file
    input_csv = 'full_histories_from_28_markets.csv'
    
    # Define the name for your output CSV file
    output_csv = 'trader_summary.csv'
    
    # Run the analysis
    trader_summary = analyze_trader_performance(input_csv)
    
    if trader_summary is not None:
        # --- EXPORT TO CSV ---
        # The new line of code is here:
        trader_summary.to_csv(output_csv, index=False)
        print(f"\nSuccessfully exported the summary to '{output_csv}'")

        # Print the results to the console as before
        print("\n--- Trader Performance Summary ---")
        print(trader_summary.to_string())
```

```python
import pandas as pd
import numpy as np

def generate_final_report(filepath):
    """
    Processes raw trade data to generate a consolidated trader report.
    """
    try:
        # Load the raw dataset
        df = pd.read_csv(filepath)
        print("Successfully loaded the raw trade history file.")
    except FileNotFoundError:
        print(f"Error: The file '{filepath}' was not found.")
        return None

    # --- 1. Initial Data Cleaning and Preparation ---
    df['Trade_Date'] = pd.to_datetime(df['Trade_Date'])
    df['Trade_Size'] = pd.to_numeric(df['Trade_Size'], errors='coerce')
    df['Trade_Price'] = pd.to_numeric(df['Trade_Price'], errors='coerce')
    df.dropna(subset=['Trade_Size', 'Trade_Price', 'Trader_Wallet', 'Trader_Name'], inplace=True)

    # --- 2. Feature Engineering ---

    # Create 'event_slug' from Market_Title
    df['event_slug'] = df['Market_Title'].str.lower().str.replace(r'[^a-z0-9\s-]', '', regex=True).str.replace(r'\s+', '-', regex=True)

    # Calculate 'date_of_joining' (first trade date for each trader)
    first_trade = df.loc[df.groupby('Trader_Wallet')['Trade_Date'].idxmin()]
    joining_dates = first_trade.set_index('Trader_Wallet')['Trade_Date']

    # --- 3. Calculate Realized P&L and Open Positions ---

    # Calculate the signed value for each trade (cost vs revenue)
    df['signed_value'] = np.where(df['Trade_Side'] == 'BUY', -df['Trade_Size'] * df['Trade_Price'], df['Trade_Size'] * df['Trade_Price'])

    # Calculate Realized P&L per trader per market
    market_pnl = df.groupby(['Trader_Wallet', 'Trader_Name', 'Market_Condition_ID', 'event_slug'])['signed_value'].sum().reset_index()
    market_pnl.rename(columns={'signed_value': 'profit_loss_total_realized'}, inplace=True)

    # Calculate Open Positions (net shares)
    buys = df[df['Trade_Side'] == 'BUY'].groupby(['Trader_Wallet', 'Market_Condition_ID'])['Trade_Size'].sum()
    sells = df[df['Trade_Side'] == 'SELL'].groupby(['Trader_Wallet', 'Market_Condition_ID'])['Trade_Size'].sum()
    positions = (buys.subtract(sells, fill_value=0)).reset_index(name='open_positions')
    
    # --- 4. Merge Data and Identify Biggest Wins/Losses ---

    # Merge P&L with open positions
    final_df = pd.merge(market_pnl, positions, on=['Trader_Wallet', 'Market_Condition_ID'], how='left')
    final_df['open_positions'].fillna(0, inplace=True)
    
    # Add the joining date
    final_df['date_of_joining'] = final_df['Trader_Wallet'].map(joining_dates)

    # Flag Biggest Win and Biggest Loss for each trader
    final_df['biggest_win'] = False
    final_df['biggest_loss'] = False

    # Get indices of max and min P&L for each trader
    max_pnl_indices = final_df.loc[final_df.groupby('Trader_Wallet')['profit_loss_total_realized'].idxmax()].index
    min_pnl_indices = final_df.loc[final_df.groupby('Trader_Wallet')['profit_loss_total_realized'].idxmin()].index

    final_df.loc[max_pnl_indices, 'biggest_win'] = True
    final_df.loc[min_pnl_indices, 'biggest_loss'] = True
    
    # --- 5. Final Formatting ---

    # Add the placeholder 'unrealized_profit_loss' column
    final_df['unrealized_profit_loss'] = 0

    # Select and reorder columns to match the requested format
    final_df = final_df[[
        'Trader_Wallet',
        'Trader_Name',
        'date_of_joining',
        'Market_Condition_ID',
        'event_slug',
        'profit_loss_total_realized',
        'open_positions',
        'unrealized_profit_loss',
        'biggest_win',
        'biggest_loss'
    ]]
    
    # Sort for better readability
    final_df = final_df.sort_values(by=['Trader_Wallet', 'profit_loss_total_realized'], ascending=[True, False])

    return final_df

if __name__ == "__main__":
    input_file = 'full_histories_from_28_markets.csv'
    output_file = 'consolidated_trader_report.csv'
    
    report = generate_final_report(input_file)
    
    if report is not None:
        # Export the final DataFrame to a CSV file
        report.to_csv(output_file, index=False)
        print(f"\nSuccessfully generated and saved the report to '{output_file}'")
        print("\nHere's a preview of the first 10 rows:")
        print(report.head(10).to_string())
```

```python
import pandas as pd
import numpy as np

def generate_report_with_value(filepath):
    """
    Processes raw trade data to generate a consolidated trader report,
    including the estimated value of open positions.
    """
    try:
        df = pd.read_csv(filepath)
        print("Successfully loaded the raw trade history file.")
    except FileNotFoundError:
        print(f"Error: The file '{filepath}' was not found.")
        return None

    # --- DIAGNOSTIC: Check Trade_Side distribution ---
    print("\n=== DIAGNOSTIC INFO ===")
    print(f"Total rows: {len(df)}")
    print(f"Trade_Side distribution:\n{df['Trade_Side'].value_counts()}")
    print(f"Sample data:\n{df[['Trader_Wallet', 'Market_Condition_ID', 'Trade_Side', 'Trade_Size', 'Trade_Price']].head(10)}")
    
    # --- 1. Data Cleaning and Preparation ---
    df['Trade_Date'] = pd.to_datetime(df['Trade_Date'])
    df['Trade_Size'] = pd.to_numeric(df['Trade_Size'], errors='coerce')
    df['Trade_Price'] = pd.to_numeric(df['Trade_Price'], errors='coerce')
    df.dropna(subset=['Trade_Size', 'Trade_Price', 'Trader_Wallet', 'Trader_Name'], inplace=True)

    # --- 2. Feature Engineering & Base Calculations ---
    df['event_slug'] = df['Market_Title'].str.lower().str.replace(r'[^a-z0-9\s-]', '', regex=True).str.replace(r'\s+', '-', regex=True)
    first_trade = df.loc[df.groupby('Trader_Wallet')['Trade_Date'].idxmin()]
    joining_dates = first_trade.set_index('Trader_Wallet')['Trade_Date']

    # --- 3. Calculate Positions and P&L (Realized vs Unrealized) ---
    
    # Find the last traded price for each market ("current price")
    last_trade_indices = df.loc[df.groupby('Market_Condition_ID')['Trade_Timestamp'].idxmax()]
    market_current_price = last_trade_indices.set_index('Market_Condition_ID')['Trade_Price']

    # Calculate comprehensive metrics per trader-market-outcome combination
    trader_market_summary = df.groupby(['Trader_Wallet', 'Trader_Name', 'Market_Condition_ID', 'Trade_Outcome']).agg(
        total_buys=('Trade_Size', lambda x: x[df.loc[x.index, 'Trade_Side'] == 'BUY'].sum()),
        total_sells=('Trade_Size', lambda x: x[df.loc[x.index, 'Trade_Side'] == 'SELL'].sum()),
        total_buy_cost=('Trade_Price', lambda x: (x[df.loc[x.index, 'Trade_Side'] == 'BUY'] * 
                                                   df.loc[x.index[df.loc[x.index, 'Trade_Side'] == 'BUY'], 'Trade_Size']).sum()),
        total_sell_proceeds=('Trade_Price', lambda x: (x[df.loc[x.index, 'Trade_Side'] == 'SELL'] * 
                                                        df.loc[x.index[df.loc[x.index, 'Trade_Side'] == 'SELL'], 'Trade_Size']).sum())
    ).reset_index()

    # --- DIAGNOSTIC: Check aggregation ---
    print(f"\nSample aggregated data:\n{trader_market_summary[['Trader_Wallet', 'Market_Condition_ID', 'Trade_Outcome', 'total_buys', 'total_sells']].head(10)}")

    # Calculate closed positions (minimum of buys and sells)
    trader_market_summary['number_of_positions_closed'] = np.minimum(
        trader_market_summary['total_buys'], 
        trader_market_summary['total_sells']
    )

    # Calculate average prices
    trader_market_summary['avg_buy_price'] = np.where(
        trader_market_summary['total_buys'] > 0,
        trader_market_summary['total_buy_cost'] / trader_market_summary['total_buys'],
        0
    )
    trader_market_summary['avg_sell_price'] = np.where(
        trader_market_summary['total_sells'] > 0,
        trader_market_summary['total_sell_proceeds'] / trader_market_summary['total_sells'],
        0
    )

    # Calculate REALIZED P&L (only from closed positions)
    trader_market_summary['profit_loss_total_realized'] = (
        trader_market_summary['number_of_positions_closed'] * 
        (trader_market_summary['avg_sell_price'] - trader_market_summary['avg_buy_price'])
    )

    # Calculate open positions
    open_positions_raw = trader_market_summary['total_buys'] - trader_market_summary['total_sells']
    trader_market_summary['open_position_direction'] = np.where(open_positions_raw > 0, 'Long', 
                                                                 np.where(open_positions_raw < 0, 'Short', 'Flat'))
    trader_market_summary['open_positions'] = open_positions_raw.abs()

    # Map current market price
    trader_market_summary['current_price'] = trader_market_summary['Market_Condition_ID'].map(market_current_price)

    # Calculate value of open positions
    trader_market_summary['value_of_open_positions'] = (
        trader_market_summary['open_positions'] * trader_market_summary['current_price']
    )

    # Calculate cost basis of open positions
    trader_market_summary['cost_of_open_positions'] = np.where(
        open_positions_raw > 0,  # Long position (bought more than sold)
        trader_market_summary['total_buy_cost'] - (trader_market_summary['number_of_positions_closed'] * trader_market_summary['avg_buy_price']),
        np.where(
            open_positions_raw < 0,  # Short position (sold more than bought)
            trader_market_summary['total_sell_proceeds'] - (trader_market_summary['number_of_positions_closed'] * trader_market_summary['avg_sell_price']),
            0  # No open positions
        )
    )

    # Calculate UNREALIZED P&L
    trader_market_summary['unrealized_profit_loss'] = np.where(
        open_positions_raw > 0,  # Long position
        trader_market_summary['value_of_open_positions'] - trader_market_summary['cost_of_open_positions'],
        np.where(
            open_positions_raw < 0,  # Short position
            trader_market_summary['cost_of_open_positions'] - trader_market_summary['value_of_open_positions'],
            0  # No open positions
        )
    )

    # --- 4. Aggregate by Market (combine all outcomes) ---
    # Create outcome_bet string
    outcome_mapping = trader_market_summary.groupby(['Trader_Wallet', 'Market_Condition_ID'])['Trade_Outcome'].apply(
        lambda x: ', '.join(x.unique())
    ).reset_index(name='outcome_bet')

    # Aggregate P&L and positions by trader-market (sum across all outcomes)
    final_summary = trader_market_summary.groupby(['Trader_Wallet', 'Trader_Name', 'Market_Condition_ID']).agg(
        profit_loss_total_realized=('profit_loss_total_realized', 'sum'),
        open_positions=('open_positions', 'sum'),
        number_of_positions_closed=('number_of_positions_closed', 'sum'),
        unrealized_profit_loss=('unrealized_profit_loss', 'sum'),
        value_of_open_positions=('value_of_open_positions', 'sum')
    ).reset_index()

    # Determine overall position direction for the market
    long_positions = trader_market_summary[trader_market_summary['open_position_direction'] == 'Long'].groupby(
        ['Trader_Wallet', 'Market_Condition_ID'])['open_positions'].sum()
    short_positions = trader_market_summary[trader_market_summary['open_position_direction'] == 'Short'].groupby(
        ['Trader_Wallet', 'Market_Condition_ID'])['open_positions'].sum()
    
    final_summary['long_total'] = final_summary.set_index(['Trader_Wallet', 'Market_Condition_ID']).index.map(long_positions).fillna(0).values
    final_summary['short_total'] = final_summary.set_index(['Trader_Wallet', 'Market_Condition_ID']).index.map(short_positions).fillna(0).values
    
    final_summary['open_position_direction'] = np.where(
        final_summary['long_total'] > final_summary['short_total'], 'Long',
        np.where(final_summary['short_total'] > final_summary['long_total'], 'Short', 'Flat')
    )
    final_summary.drop(['long_total', 'short_total'], axis=1, inplace=True)

    # --- 5. Add event_slug and outcomes ---
    event_slugs = df.groupby('Market_Condition_ID')['event_slug'].first()
    final_summary['event_slug'] = final_summary['Market_Condition_ID'].map(event_slugs)
    
    final_summary = pd.merge(final_summary, outcome_mapping, on=['Trader_Wallet', 'Market_Condition_ID'], how='left')

    # --- 6. Add date_of_joining ---
    final_summary['date_of_joining'] = final_summary['Trader_Wallet'].map(joining_dates)

    # --- 7. Flag Biggest Win/Loss per trader ---
    final_summary['biggest_win'] = False
    final_summary['biggest_loss'] = False
    
    # Only flag if there's actual realized P&L
    traders_with_pnl = final_summary[final_summary['profit_loss_total_realized'] != 0].groupby('Trader_Wallet')
    
    if len(traders_with_pnl) > 0:
        max_pnl_indices = final_summary.loc[
            final_summary[final_summary['profit_loss_total_realized'] != 0]
            .groupby('Trader_Wallet')['profit_loss_total_realized'].idxmax()
        ].index
        min_pnl_indices = final_summary.loc[
            final_summary[final_summary['profit_loss_total_realized'] != 0]
            .groupby('Trader_Wallet')['profit_loss_total_realized'].idxmin()
        ].index
        final_summary.loc[max_pnl_indices, 'biggest_win'] = True
        final_summary.loc[min_pnl_indices, 'biggest_loss'] = True

    # --- 8. Final Formatting ---
    final_summary.fillna(0, inplace=True)
    
    # Round numerical columns
    cols_to_round = ['profit_loss_total_realized', 'open_positions', 'number_of_positions_closed', 
                     'unrealized_profit_loss', 'value_of_open_positions']
    final_summary[cols_to_round] = final_summary[cols_to_round].round(2)
    
    # Select and reorder columns
    final_df = final_summary[[
        'Trader_Wallet', 'Trader_Name', 'date_of_joining', 'Market_Condition_ID',
        'event_slug', 'outcome_bet', 'profit_loss_total_realized', 'open_positions', 
        'value_of_open_positions', 'open_position_direction', 
        'number_of_positions_closed', 'unrealized_profit_loss',
        'biggest_win', 'biggest_loss'
    ]]
    
    final_df = final_df.sort_values(by=['Trader_Wallet', 'profit_loss_total_realized'], ascending=[True, False])

    # --- DIAGNOSTIC: Final check ---
    print(f"\nFinal summary - positions closed distribution:")
    print(final_df['number_of_positions_closed'].describe())
    print(f"\nSample final data:\n{final_df[['Trader_Wallet', 'Market_Condition_ID', 'profit_loss_total_realized', 'number_of_positions_closed']].head(10)}")

    return final_df

if __name__ == "__main__":
    input_file = 'full_histories_from_28_markets.csv'
    output_file = 'final_trader_report_with_value.csv'
    
    report = generate_report_with_value(input_file)
    
    if report is not None:
        report.to_csv(output_file, index=False)
        print(f"\nSuccessfully generated and saved the new report to '{output_file}'")
        print("\nHere's a preview:")
        print(report.head(10).to_string())
```

```python
import pandas as pd
import numpy as np

def generate_comprehensive_trader_report(filepath):
    """
    Generate a comprehensive trader report showing realized and unrealized P&L
    across all markets, matching Polymarket's trader profile calculations.
    """
    try:
        df = pd.read_csv(filepath)
        print(f"Successfully loaded {len(df)} trade records.")
    except FileNotFoundError:
        print(f"Error: The file '{filepath}' was not found.")
        return None

    # --- 1. Data Cleaning ---
    df['Trade_Date'] = pd.to_datetime(df['Trade_Date'])
    df['Trade_Size'] = pd.to_numeric(df['Trade_Size'], errors='coerce')
    df['Trade_Price'] = pd.to_numeric(df['Trade_Price'], errors='coerce')
    df.dropna(subset=['Trade_Size', 'Trade_Price', 'Trader_Wallet', 'Trader_Name'], inplace=True)

    print(f"\n=== DATA OVERVIEW ===")
    print(f"Total traders: {df['Trader_Wallet'].nunique()}")
    print(f"Total markets: {df['Market_Condition_ID'].nunique()}")
    print(f"Date range: {df['Trade_Date'].min()} to {df['Trade_Date'].max()}")

    # --- 2. Calculate Date of Joining per Trader ---
    first_trade = df.loc[df.groupby('Trader_Wallet')['Trade_Date'].idxmin()]
    joining_dates = first_trade[['Trader_Wallet', 'Trade_Date', 'Trader_Name']].rename(
        columns={'Trade_Date': 'date_of_joining'}
    )

    # --- 3. Create Event Slug ---
    df['event_slug'] = df['Market_Title'].str.lower().str.replace(
        r'[^a-z0-9\s-]', '', regex=True
    ).str.replace(r'\s+', '-', regex=True)

    # --- 4. Calculate Per-Market Metrics ---
    trader_market_metrics = []

    for (wallet, name, market_id), group in df.groupby(['Trader_Wallet', 'Trader_Name', 'Market_Condition_ID']):
        # Get market info
        event_slug = group['event_slug'].iloc[0]
        outcomes_traded = ', '.join(sorted(group['Trade_Outcome'].unique()))
        
        # Separate buys and sells
        buys = group[group['Trade_Side'] == 'BUY']
        sells = group[group['Trade_Side'] == 'SELL']
        
        total_buy_shares = buys['Trade_Size'].sum()
        total_sell_shares = sells['Trade_Size'].sum()
        
        total_buy_cost = (buys['Trade_Size'] * buys['Trade_Price']).sum()
        total_sell_proceeds = (sells['Trade_Size'] * sells['Trade_Price']).sum()
        
        # Calculate closed positions
        num_closed = min(total_buy_shares, total_sell_shares)
        
        # Calculate average prices
        avg_buy_price = total_buy_cost / total_buy_shares if total_buy_shares > 0 else 0
        avg_sell_price = total_sell_proceeds / total_sell_shares if total_sell_shares > 0 else 0
        
        # REALIZED P&L (from closed positions only)
        realized_pnl = (avg_sell_price - avg_buy_price) * num_closed
        
        # Calculate open positions
        net_shares = total_buy_shares - total_sell_shares
        open_shares = abs(net_shares)
        position_direction = 'Long' if net_shares > 0 else ('Short' if net_shares < 0 else 'Flat')
        
        # Get current price (last trade in this market)
        current_price = group.loc[group['Trade_Timestamp'].idxmax(), 'Trade_Price']
        
        # Value of open positions
        value_open = open_shares * current_price
        
        # UNREALIZED P&L
        if net_shares > 0:  # Long position
            cost_of_open = total_buy_cost - (num_closed * avg_buy_price)
            unrealized_pnl = (open_shares * current_price) - cost_of_open
        elif net_shares < 0:  # Short position
            proceeds_from_open = total_sell_proceeds - (num_closed * avg_sell_price)
            unrealized_pnl = proceeds_from_open - (open_shares * current_price)
        else:  # Flat
            unrealized_pnl = 0
        
        trader_market_metrics.append({
            'trader_wallet': wallet,
            'trader_name': name,
            'market_condition_id': market_id,
            'event_slug': event_slug,
            'outcome_bet': outcomes_traded,
            'profit_loss_total_realized': round(realized_pnl, 2),
            'open_positions': round(open_shares, 2),
            'value_of_open_positions': round(value_open, 2),
            'open_position_direction': position_direction,
            'number_of_positions_closed': round(num_closed, 2),
            'unrealized_profit_loss': round(unrealized_pnl, 2),
            'total_pnl': round(realized_pnl + unrealized_pnl, 2)
        })

    result_df = pd.DataFrame(trader_market_metrics)
    
    # --- 5. Add Date of Joining ---
    result_df = result_df.merge(joining_dates, left_on='trader_wallet', right_on='Trader_Wallet', how='left')
    result_df.drop(columns=['Trader_Wallet'], inplace=True)
    
    # --- 6. Flag Biggest Win/Loss per Trader ---
    result_df['biggest_win'] = False
    result_df['biggest_loss'] = False
    
    for trader in result_df['trader_wallet'].unique():
        trader_data = result_df[result_df['trader_wallet'] == trader]
        
        # Biggest win based on realized P&L
        if len(trader_data[trader_data['profit_loss_total_realized'] > 0]) > 0:
            max_idx = trader_data['profit_loss_total_realized'].idxmax()
            result_df.loc[max_idx, 'biggest_win'] = True
        
        # Biggest loss based on realized P&L
        if len(trader_data[trader_data['profit_loss_total_realized'] < 0]) > 0:
            min_idx = trader_data['profit_loss_total_realized'].idxmin()
            result_df.loc[min_idx, 'biggest_loss'] = True

    # Reorder columns to match requirements
    result_df = result_df[['trader_wallet', 'trader_name', 'date_of_joining', 'market_condition_id', 
                            'event_slug', 'outcome_bet', 'profit_loss_total_realized', 'open_positions', 
                            'value_of_open_positions', 'open_position_direction', 'number_of_positions_closed',
                            'unrealized_profit_loss', 'biggest_win', 'biggest_loss']]

    # --- 7. Sort by trader and P&L ---
    result_df = result_df.sort_values(
        by=['trader_wallet', 'profit_loss_total_realized'], 
        ascending=[True, False]
    )

    # --- 8. Create Summary Statistics ---
    print(f"\n=== REPORT SUMMARY ===")
    print(f"Total trader-market combinations: {len(result_df)}")
    print(f"\nRealized P&L Distribution:")
    print(result_df['profit_loss_total_realized'].describe())
    print(f"\nTraders with closed positions: {len(result_df[result_df['number_of_positions_closed'] > 0])}")
    print(f"Traders with only open positions: {len(result_df[result_df['number_of_positions_closed'] == 0])}")

    return result_df


def generate_trader_summary(report_df):
    """
    Generate a trader-level summary (like Polymarket profile page)
    showing total account performance across all markets.
    """
    trader_summary = report_df.groupby(['trader_wallet', 'trader_name', 'date_of_joining']).agg({
        'profit_loss_total_realized': 'sum',
        'unrealized_profit_loss': 'sum',
        'value_of_open_positions': 'sum',
        'market_condition_id': 'count'
    }).reset_index()

    trader_summary.rename(columns={'market_condition_id': 'number_of_predictions'}, inplace=True)
    trader_summary['total_account_pnl'] = (
        trader_summary['profit_loss_total_realized'] + 
        trader_summary['unrealized_profit_loss']
    )
    
    # Round values
    trader_summary['profit_loss_total_realized'] = trader_summary['profit_loss_total_realized'].round(2)
    trader_summary['unrealized_profit_loss'] = trader_summary['unrealized_profit_loss'].round(2)
    trader_summary['value_of_open_positions'] = trader_summary['value_of_open_positions'].round(2)
    trader_summary['total_account_pnl'] = trader_summary['total_account_pnl'].round(2)
    
    # Sort by total P&L
    trader_summary = trader_summary.sort_values(by='total_account_pnl', ascending=False)
    
    return trader_summary


if __name__ == "__main__":
    input_file = 'full_histories_from_28_markets.csv'
    output_file_detailed = 'final_trader_report_detailed.csv'
    output_file_summary = 'trader_summary_report.csv'
    
    # Generate detailed report (per market)
    report = generate_comprehensive_trader_report(input_file)
    
    if report is not None:
        # Save detailed report
        report.to_csv(output_file_detailed, index=False)
        print(f"\n‚úÖ Successfully saved detailed report to '{output_file_detailed}'")
        
        # Generate and save trader summary
        summary = generate_trader_summary(report)
        summary.to_csv(output_file_summary, index=False)
        print(f"‚úÖ Successfully saved trader summary to '{output_file_summary}'")
        
        print("\nüìä DETAILED REPORT SAMPLE (Per Market):")
        print(report.head(15).to_string())
        
        print("\n\nüìà TRADER SUMMARY SAMPLE (Aggregated by Trader):")
        print(summary.head(10).to_string())
        
        print(f"\n\n=== KEY INSIGHTS ===")
        print(f"Top performing trader: {summary.iloc[0]['trader_name']} with ${summary.iloc[0]['total_account_pnl']:.2f} total P&L")
        print(f"Total markets tracked: {report['market_condition_id'].nunique()}")
        print(f"Total unique traders: {report['trader_wallet'].nunique()}")
```

```python
import pandas as pd
import numpy as np

def generate_comprehensive_trader_report(filepath):
    """
    Generate a comprehensive trader report from CSV file showing realized and unrealized P&L
    across all markets for traders who participated in earnings markets.
    """
    try:
        df = pd.read_csv(filepath)
        print(f"‚úÖ Successfully loaded {len(df)} trade records.")
    except FileNotFoundError:
        print(f"‚ùå Error: The file '{filepath}' was not found.")
        return None, None

    # --- 1. Data Cleaning ---
    df['Trade_Date'] = pd.to_datetime(df['Trade_Date'])
    df['Trade_Size'] = pd.to_numeric(df['Trade_Size'], errors='coerce')
    df['Trade_Price'] = pd.to_numeric(df['Trade_Price'], errors='coerce')
    df.dropna(subset=['Trade_Size', 'Trade_Price', 'Trader_Wallet', 'Trader_Name'], inplace=True)

    print(f"\n{'='*80}")
    print("DATA OVERVIEW")
    print(f"{'='*80}")
    print(f"Total traders: {df['Trader_Wallet'].nunique()}")
    print(f"Total markets: {df['Market_Condition_ID'].nunique()}")
    print(f"Date range: {df['Trade_Date'].min().date()} to {df['Trade_Date'].max().date()}")
    print(f"Total BUY trades: {len(df[df['Trade_Side'] == 'BUY'])}")
    print(f"Total SELL trades: {len(df[df['Trade_Side'] == 'SELL'])}")

    # --- 2. Calculate Date of Joining per Trader ---
    first_trade = df.loc[df.groupby('Trader_Wallet')['Trade_Date'].idxmin()]
    joining_dates = first_trade[['Trader_Wallet', 'Trade_Date', 'Trader_Name']].rename(
        columns={'Trade_Date': 'date_of_joining'}
    )

    # --- 3. Create Event Slug ---
    df['event_slug'] = df['Market_Title'].str.lower().str.replace(
        r'[^a-z0-9\s-]', '', regex=True
    ).str.replace(r'\s+', '-', regex=True)

    # --- 4. Calculate Per-Market Metrics ---
    print(f"\n{'='*80}")
    print("CALCULATING TRADER METRICS")
    print(f"{'='*80}")
    
    trader_market_metrics = []
    
    grouped = df.groupby(['Trader_Wallet', 'Trader_Name', 'Market_Condition_ID'])
    total_groups = len(grouped)
    print(f"Processing {total_groups} trader-market combinations...\n")

    for idx, ((wallet, name, market_id), group) in enumerate(grouped, 1):
        if idx % 100 == 0:
            print(f"  Progress: {idx}/{total_groups} ({idx/total_groups*100:.1f}%)")
        
        # Get market info
        event_slug = group['event_slug'].iloc[0]
        outcomes_traded = ', '.join(sorted(group['Trade_Outcome'].unique()))
        
        # Separate buys and sells
        buys = group[group['Trade_Side'] == 'BUY']
        sells = group[group['Trade_Side'] == 'SELL']
        
        total_buy_shares = buys['Trade_Size'].sum()
        total_sell_shares = sells['Trade_Size'].sum()
        
        total_buy_cost = (buys['Trade_Size'] * buys['Trade_Price']).sum()
        total_sell_proceeds = (sells['Trade_Size'] * sells['Trade_Price']).sum()
        
        # Calculate closed positions
        num_closed = min(total_buy_shares, total_sell_shares)
        
        # Calculate average prices
        avg_buy_price = total_buy_cost / total_buy_shares if total_buy_shares > 0 else 0
        avg_sell_price = total_sell_proceeds / total_sell_shares if total_sell_shares > 0 else 0
        
        # REALIZED P&L (from closed positions only)
        realized_pnl = (avg_sell_price - avg_buy_price) * num_closed
        
        # Calculate open positions
        net_shares = total_buy_shares - total_sell_shares
        open_shares = abs(net_shares)
        position_direction = 'Long' if net_shares > 0 else ('Short' if net_shares < 0 else 'Flat')
        
        # Get current price (last trade in this market)
        current_price = group.loc[group['Trade_Timestamp'].idxmax(), 'Trade_Price']
        
        # Value of open positions
        value_open = open_shares * current_price
        
        # UNREALIZED P&L
        if net_shares > 0:  # Long position
            cost_of_open = total_buy_cost - (num_closed * avg_buy_price)
            unrealized_pnl = (open_shares * current_price) - cost_of_open
        elif net_shares < 0:  # Short position
            proceeds_from_open = total_sell_proceeds - (num_closed * avg_sell_price)
            unrealized_pnl = proceeds_from_open - (open_shares * current_price)
        else:  # Flat
            unrealized_pnl = 0
        
        trader_market_metrics.append({
            'trader_wallet': wallet,
            'trader_name': name,
            'market_condition_id': market_id,
            'event_slug': event_slug,
            'outcome_bet': outcomes_traded,
            'profit_loss_total_realized': round(realized_pnl, 2),
            'open_positions': round(open_shares, 2),
            'value_of_open_positions': round(value_open, 2),
            'open_position_direction': position_direction,
            'number_of_positions_closed': round(num_closed, 2),
            'unrealized_profit_loss': round(unrealized_pnl, 2),
            'total_pnl': round(realized_pnl + unrealized_pnl, 2)
        })

    result_df = pd.DataFrame(trader_market_metrics)
    
    # --- 5. Add Date of Joining ---
    result_df = result_df.merge(joining_dates, left_on='trader_wallet', right_on='Trader_Wallet', how='left')
    result_df.drop(columns=['Trader_Wallet'], inplace=True)
    
    # --- 6. Flag Biggest Win/Loss per Trader ---
    result_df['biggest_win'] = False
    result_df['biggest_loss'] = False
    
    for wallet in result_df['trader_wallet'].unique():
        trader_data = result_df[result_df['trader_wallet'] == wallet]
        
        # Biggest win based on realized P&L
        if len(trader_data[trader_data['profit_loss_total_realized'] > 0]) > 0:
            max_idx = trader_data['profit_loss_total_realized'].idxmax()
            result_df.loc[max_idx, 'biggest_win'] = True
        
        # Biggest loss based on realized P&L
        if len(trader_data[trader_data['profit_loss_total_realized'] < 0]) > 0:
            min_idx = trader_data['profit_loss_total_realized'].idxmin()
            result_df.loc[min_idx, 'biggest_loss'] = True

    # Reorder columns to match requirements
    result_df = result_df[['trader_wallet', 'trader_name', 'date_of_joining', 'market_condition_id', 
                            'event_slug', 'outcome_bet', 'profit_loss_total_realized', 'open_positions', 
                            'value_of_open_positions', 'open_position_direction', 'number_of_positions_closed',
                            'unrealized_profit_loss', 'biggest_win', 'biggest_loss']]

    # --- 7. Sort by trader and P&L ---
    result_df = result_df.sort_values(
        by=['trader_wallet', 'profit_loss_total_realized'], 
        ascending=[True, False]
    )

    # --- 8. Generate Summary Statistics ---
    print(f"\n{'='*80}")
    print("DETAILED REPORT SUMMARY")
    print(f"{'='*80}")
    print(f"Total trader-market combinations: {len(result_df)}")
    print(f"Traders with closed positions: {len(result_df[result_df['number_of_positions_closed'] > 0])}")
    print(f"Traders with only open positions: {len(result_df[result_df['number_of_positions_closed'] == 0])}")
    print(f"\nRealized P&L Distribution:")
    print(result_df['profit_loss_total_realized'].describe())

    return result_df, joining_dates


def generate_trader_summary(report_df):
    """
    Generate a trader-level summary (like Polymarket profile page)
    showing total account performance across all markets.
    """
    trader_summary = report_df.groupby(['trader_wallet', 'trader_name', 'date_of_joining']).agg({
        'profit_loss_total_realized': 'sum',
        'unrealized_profit_loss': 'sum',
        'value_of_open_positions': 'sum',
        'market_condition_id': 'count'
    }).reset_index()

    trader_summary.rename(columns={'market_condition_id': 'number_of_predictions'}, inplace=True)
    trader_summary['total_account_pnl'] = (
        trader_summary['profit_loss_total_realized'] + 
        trader_summary['unrealized_profit_loss']
    )
    
    # Round values
    trader_summary['profit_loss_total_realized'] = trader_summary['profit_loss_total_realized'].round(2)
    trader_summary['unrealized_profit_loss'] = trader_summary['unrealized_profit_loss'].round(2)
    trader_summary['value_of_open_positions'] = trader_summary['value_of_open_positions'].round(2)
    trader_summary['total_account_pnl'] = trader_summary['total_account_pnl'].round(2)
    
    # Sort by total P&L
    trader_summary = trader_summary.sort_values(by='total_account_pnl', ascending=False)
    
    return trader_summary


if __name__ == "__main__":
    input_file = 'full_histories_from_28_markets.csv'
    output_file_detailed = 'earnings_traders_detailed_report.csv'
    output_file_summary = 'earnings_traders_summary.csv'
    
    print(f"{'='*80}")
    print("EARNINGS MARKET TRADER REPORT GENERATOR")
    print(f"Current Time: 2025-10-02 16:32:03 UTC")
    print(f"{'='*80}\n")
    
    # Generate detailed report (per market)
    report, joining_dates = generate_comprehensive_trader_report(input_file)
    
    if report is not None:
        # Save detailed report
        report.to_csv(output_file_detailed, index=False)
        print(f"\n‚úÖ Successfully saved detailed report to '{output_file_detailed}'")
        
        # Generate and save trader summary
        summary = generate_trader_summary(report)
        summary.to_csv(output_file_summary, index=False)
        print(f"‚úÖ Successfully saved trader summary to '{output_file_summary}'")
        
        print(f"\n{'='*80}")
        print("DETAILED REPORT SAMPLE (Per Market)")
        print(f"{'='*80}")
        print(report.head(20).to_string())
        
        print(f"\n{'='*80}")
        print("TRADER SUMMARY SAMPLE (Aggregated by Trader)")
        print(f"{'='*80}")
        print(summary.head(15).to_string())
        
        print(f"\n{'='*80}")
        print("KEY INSIGHTS")
        print(f"{'='*80}")
        if len(summary) > 0:
            print(f"üèÜ Top trader: {summary.iloc[0]['trader_name']}")
            print(f"   Total P&L: ${summary.iloc[0]['total_account_pnl']:.2f}")
            print(f"   Markets traded: {summary.iloc[0]['number_of_predictions']}")
            print(f"\nüìä Average P&L per trader: ${summary['total_account_pnl'].mean():.2f}")
            print(f"üìä Median P&L per trader: ${summary['total_account_pnl'].median():.2f}")
            print(f"üìä Total unique traders: {len(summary)}")
            print(f"üìä Total markets: {report['market_condition_id'].nunique()}")
```
